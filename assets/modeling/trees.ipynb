```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import statsmodels.api as sm
```

```python
trees = pd.read_csv('https://forge.scilab.org/index.php/p/rdataset/source/file/master/csv/datasets/trees.csv')
trees=trees.drop(['Unnamed: 0', 'Girth'], axis=1)
print(trees.head(10).to_markdown())
```

```python
trees.columns
```

```python
plt.scatter(trees['Height'],trees['Volume'])
plt.xlabel('Height (ft.)')
plt.ylabel('Volume (ft.^3)')
plt.savefig("treeplot.jpg")
plt.show()
```

```python
plt.scatter(trees['Height'],trees['Volume'])
plt.xlabel('Height (ft.)')
plt.ylabel('Volume (ft.^3)')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
plt.plot(x,2*x-120, 'r-', lw=4)
font = {'family': 'serif',
        'color':  'red',
        'weight': 'normal',
        'size': 40,
        }
plt.text(67,55, "?", font)
plt.savefig("assumption.jpg")
plt.show()
```

```python
plt.scatter(trees['Height'],trees['Volume'])
plt.xlabel('Height (ft.)')
plt.ylabel('Volume (ft.^3)')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax)
plt.plot(x,2*x-120, 'r-', lw=4)
plt.arrow(85.6,2*85.6-120, 0, -20, lw=3, color='purple', head_width=1, head_length=5)
plt.arrow(85.6,2*85.6-120, 0, 20, lw=3, color='purple', head_width=1, head_length=5)
font = {'family': 'serif',
        'color':  'red',
        'weight': 'normal',
        'size': 40,
        }
#plt.text(67,55, "?", font)
plt.savefig("confidence.jpg")
plt.show()
```

```python
out = sm.OLS(trees['Height'],trees['Volume'])
result = out.fit()
print(result.summary())
```
